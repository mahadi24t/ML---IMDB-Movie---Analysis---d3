```python


# Create subplots with 1 row and 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2)

# First subplot - Rank vs Metascore
ax1.plot(df['Rank'], df['Metascore'])
ax1.set_xlim(1, 20)
ax1.set_ylim(0, 100)
ax1.set_xlabel('Rank')
ax1.set_ylabel('Metascore')
ax1.set_title('Rank vs Metascore')
ax1.grid(True)

# Second subplot - Rank vs Rating
ax2.plot(df['Rank'], df['Rating'])
ax2.set_xlim(1, 20)
ax2.set_ylim(0, 10)
ax2.set_xlabel('Rank')
ax2.set_ylabel('Rating')
ax2.set_title('Rank vs Rating')
ax2.grid(True)

# Adjust spacing between subplots
plt.tight_layout()



# Display the chart
plt.show()
```

```python


# Filter the DataFrame for movies with rating between 8 and 10
filtered_df = df[(df['Rating'] >= 8.5) & (df['Rating'] <= 10)]

# Group the filtered DataFrame by 'Director' and count the number of movies for each director
director_counts = filtered_df['Director'].value_counts()

# Create a pie chart
plt.pie(director_counts, labels=director_counts.index, autopct='%1.1f%%')

# Set title
plt.title('Movie Director Distribution (Rating: 8.5-10)')

# Display the chart
plt.show()


```

```python
total_genre_count = df['Genre'].nunique()
total_genre_count
```

```python
# Calculate the total number of votes for each genre
genre_votes = df.groupby('Genre')['Votes'].sum().sort_values(ascending=False)

# Convert the total votes to millions
genre_votes_millions = genre_votes / 1000000

# Select the top 15 genres based on total votes
top_genres = genre_votes_millions.head(15)

# Create a bar chart to visualize the genre popularity
plt.bar(top_genres.index, top_genres)

# Set labels and title
plt.xlabel('Genre')
plt.ylabel('Total Votes (Millions)')
plt.title('Top 15 Movie Genres by Viewer Counts')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Display the chart
plt.show()

```
